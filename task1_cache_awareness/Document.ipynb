We observe some overview points:
- If the working set fits into L1 Cache, which represents the case with working set as 1024 Bytes, the 3 access patterns do not differ much as aspect of runtime.
- The *Sequential Access* case and *Reverse Sequential Access* case have similar performance. But the *Sequential Access* case is always better.
- As the stride size grows, the performance decreses in case of both sequential cases.
- If the working set is beyond L2 Cache and even the stride size is larger than L2 Cache, the performance of both sequential cases is worse than the *Random Access* case.

## Sequential Read Access Pattern

````python
df_i7_seq    = pd.read_csv("i7-8550u/sequential.csv")
df_i7_seq8 = df_i7_seq[df_i7_seq['stride(Bytes)'] == 8]
workingset   = df_i7_seq8['workingset(Bytes)']
cycles8       = df_i7_seq8['cycles']
cycles8       = [float(item) for item in cycles8]

df_i7_seq64 = df_i7_seq[df_i7_seq['stride(Bytes)'] == 64]
cycles64      = df_i7_seq64['cycles']
cycles64      = [float(item) for item in cycles64]

df_i7_seq128 = df_i7_seq[df_i7_seq['stride(Bytes)'] == 128]
cycles128      = df_i7_seq128['cycles']
cycles128      = [float(item) for item in cycles128]

df_i7_seq256 = df_i7_seq[df_i7_seq['stride(Bytes)'] == 256]
cycles256      = df_i7_seq256['cycles']
cycles256      = [float(item) for item in cycles256]

df_i7_seq32768 = df_i7_seq[df_i7_seq['stride(Bytes)'] == 32768]
cycles32768      = df_i7_seq32768['cycles']
cycles32768      = [float(item) for item in cycles32768]
workingset32768    = df_i7_seq32768['workingset(Bytes)']

for i in range(20):
    cycles8[i] = cycles8[i] / (workingset[i] / 8)
    cycles64[i] = cycles64[i] / (workingset[i] / 8)    
    cycles128[i] = cycles128[i] / (workingset[i] / 8) 
    cycles256[i] = cycles256[i] / (workingset[i] / 8)

for i in range(10): 
    cycles32768[i] = cycles32768[i] / (workingset32768[80 + i] / 8.0)
    
for i in range(20):
    workingset[i] = math.log2(workingset[i])
workingset   = ["2^" + str(item) for item in workingset]

for i in range(10):
    workingset32768[80 + i] = math.log2(workingset32768[80 + i])
workingset32768   = ["2^" + str(item) for item in workingset32768]

ax = plt.gca()
ax.grid(True)
plt.tick_params(axis='x', which='major', labelsize=8)
plt.tight_layout()
plt.plot(workingset,cycles8,'ro-',workingset,cycles64,'g+-', workingset,cycles128,'b^-', workingset,cycles256,'p--', workingset32768, cycles32768, 'y+-')

plt.title('Sequential Read Access Pattern for Serveral Sizes on I7-8550U', fontsize=30, fontweight='bold')
plt.xlabel('Working Set Size (Bytes)', fontsize=20, fontweight='bold')
plt.ylabel('Cycles/List Element', fontsize=20, fontweight='bold')
plt.legend(['Stride 8 Bytes', 'Stride 64 Bytes', 'Stride 128 Byte', 'Stride 256 Bytes', 'Stride 32KiBytes'], loc='upper left', fontsize=20)
plt.figure(figsize=(400,200))
plt.show()
````

# Answer to Question 

- What observation do you make?
  1. If the working set fits into L1 Cache, the 3 access patterns do not differ much.
  2. The *Sequential Access* case and *Reverse Sequential Access* case have similar performance. But the *Sequential Access* case is always faster or they are equally fast.
  3. As the stride size grows, the performance decreses in case of both sequential cases.
- Does the experiment match your expectations / hypothesis?
  - Yes. Before the mircobenchmarking I did the google benchmark in the Introduction part. It provides throughput and runtime reflect our plots. 
  - But there are still places not matching my expectations, I append it into to last part -- My Question. In generall, I did benchmark on two CPUS - one is laptop CPU but more modern, the another is server CPU but less modern. On both CPUs, I found it hard to record the case fits into L1 cache. Maybe it is relevant with code quality, benchmarking approach or the enviorment noise.
- How can you explain the observed behavior?
  1. If the working data set fits into L1 cache, then each access is achieved within the L1 cache. Each access's latency resulted by register or L1 cache. So the access patterns do not influence the runtime, when the number of accesses remains same.
  2. In generall, the approaches are just mirroring each other. But the spatial locality assume the future accesses on higher address in address space. So most cases of prefetch are loading the *next* elements, not *previous* elements. Even in each cache-line fetch, it is more likely to load the accesses element at higher part of cache line, in order to piggyback more possible *next* elements in the lower part of cache line. So the reverse sequential pattern doesn't match the assumption of spatial locality, it is not superise to see more cache-miss in this case. 
  3. If stride size is 64Bytes, the next accessing element is not loading the current cache line. The prefetch logic can't load a new cache line every cycle. There a stall to load from L2 is planned in every iteration. With the stride size increasing, the TLB cache misses can be involved, even page misses. The performance is not only influenced by the cache, when the next access in beyond the L3 cache. In this case, each next access must suffer from memory latency, while prefetching and cache not helpful at all.
- What conclusions can be drawn from your hardware based on the results?
  - From the runtime *stages*, the size of each level cache can be inferred. The boundary, where the next *stage* begins, is, where the previous level cache ends.
  - Also from the runtime *stages*, the latency of each level cache can be inferred. Notice *the inferred latency* is a optimized case with helping from prefetching. (But in my plots reflect this information badly.) 
  - It is also to check if the CPU has prefetching feature or is prefetching enabled.
  - The memory latency can be inferred, if the stride is larger than the size of L3 cache, which means each access is followed by a memory access without any cache effect.
  - The TLB size, the OS page size can be also inferred. They are more advanced stuff and have to combine all pre-information together.
 
Indeed, the answers to the questions are described in What Every Programmer Should Know About Memory - Ulrich Drepper.
